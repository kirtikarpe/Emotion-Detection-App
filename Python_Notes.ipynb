Classes -  template for creating records(objects)

```python
job_title = "Data Analyst"
job_location = "India"
job_salary = 500000

# to store everything in one class

class JobPost:
    def __init__(self, title, location, salary):
        self.title = title
        self.location = location
        self.salary = salary
```

```python
JobPost(job_title, job_location, job_salary)
```

<font size="5"> String Formating

<font size="4">Method 1 

```python
role = "Data Analyst"
skill = "Python"
```

```python
# to write - Role: Data Analyst

"Role: {}".format(role)
```

```python
# to write - Role: Data Analyst; Skill required: Python

"Role: {}; Skill Required: {}".format(role,skill)
```

<font size="4">Method 2

```python
# F-Strings

f"Role: {role}; Skill Reguired: {skill}"   
# easy to ready and understand as compared to format method
```

<font size="4">Method 3

```python
# % formatting

"Role: %s; skill Required: %s" % (role, skill)   
# to complicate to understand  
```
